# Report: Analysis of Immigration Trends to Canada

**Introduction:**
The purpose of this report is to analyze immigration trends to Canada based on the provided dataset. The dataset consists of immigration data from 1980 to 2013, including information such as the country of origin, continent, region, and total number of immigrants.

**1. Total Immigration Trends:**
The total number of immigrants to Canada has increased over the years, with some fluctuations. From 1980 to 2013, there has been a general upward trend, with a few notable spikes and dips in between.

**2. Immigration by Continent:**
The majority of immigrants to Canada come from Asia, followed by Europe and then Africa. The least number of immigrants come from Oceania. This indicates that Canada receives immigrants from a diverse range of continents.

**3. Top 5 Countries with Highest Immigration:**
The top 5 countries with the highest immigration to Canada are shown to be consistent over the years. These countries include China, India, Philippines, Pakistan, and United States. China and India particularly stand out as the highest contributors to immigration.

**4. Immigration Trends by Development Status:**
Immigration trends vary based on the development status of countries. Developed regions generally contribute more immigrants compared to developing regions. This suggests that factors such as economic opportunities and political stability influence immigration patterns.

**5. Immigration Patterns in 2013:**
In 2013, the top 10 countries contributing to immigration to Canada were analyzed. The analysis revealed that countries like China, India, and the Philippines remained among the highest contributors.

**6. Conclusion:**
In conclusion, Canada continues to attract immigrants from various parts of the world, with Asia being the largest contributor. The immigration trends indicate a preference for Canada among both developed and developing countries. Understanding these trends is crucial for policy-makers and stakeholders involved in immigration planning and integration efforts.

**Recommendations:**
Based on the analysis, it is recommended to further explore the factors driving immigration from specific regions and countries. Additionally, ongoing monitoring and analysis of immigration trends will facilitate informed decision-making in immigration policies and programs.

Overall, this report provides valuable insights into immigration patterns to Canada, contributing to a better understanding of Canada's multicultural landscape and its implications for society and the economy.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

df=pd.read_csv("/kaggle/input/immigration-to-canada/canadian_immegration_data.csv")
```

```python
# Total Immigration by Year
total_immigration_by_year = df.loc[:, '1980':'2013'].sum()
total_immigration_by_year.plot(kind='line', figsize=(10, 6))
plt.title('Total Immigration to Canada (1980 - 2013)')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.grid(True)
plt.show()
```

```python
# Immigration by Continent
immigration_by_continent = df.groupby('Continent')['Total'].sum()
immigration_by_continent.plot(kind='bar', figsize=(10, 6))
plt.title('Immigration to Canada by Continent')
plt.xlabel('Continent')
plt.ylabel('Number of Immigrants')
plt.xticks(rotation=90)
plt.show()
```

```python
# Top 5 Countries with Highest Immigration
top_5_countries = df.sort_values(by='Total', ascending=False).head(5)
top_5_countries = top_5_countries.set_index('Country').loc[:, '1980':'2013'].transpose()

top_5_countries.plot(kind='line', figsize=(10, 6))
plt.title('Immigration Trend of Top 5 Countries to Canada (1980 - 2013)')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.grid(True)
plt.show()
```

```python
# Boxplot of Immigration from Asia
immigration_asia = df[df['Continent'] == 'Asia'].loc[:, '1980':'2013']
immigration_asia.transpose().plot(kind='box', figsize=(10, 6))
plt.title('Boxplot of Immigration from Asia to Canada (1980 - 2013)')
plt.ylabel('Number of Immigrants')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
# Heatmap of Immigration by Region
immigration_by_region = df.pivot_table(index='Region', columns='Country', values='Total', aggfunc='sum')
plt.figure(figsize=(12, 8))
sns.heatmap(immigration_by_region, cmap='YlGnBu')
plt.title('Heatmap of Immigration to Canada by Region (1980 - 2013)')
plt.xlabel('Country')
plt.ylabel('Region')
plt.show()
```

```python
# Stacked Area Chart of Immigration by Development Status
immigration_by_dev_status = df.groupby('DevName').sum().loc[:, '1980':'2013']
immigration_by_dev_status.transpose().plot(kind='area', stacked=True, figsize=(10, 6))
plt.title('Stacked Area Chart of Immigration to Canada by Development Status (1980 - 2013)')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.grid(True)
plt.show()
```

```python
# Bar Chart of Immigration in 2013 by Top 10 Countries
immigration_2013_top_10 = df.sort_values(by='2013', ascending=False).head(10)
immigration_2013_top_10.plot(kind='bar', x='Country', y='2013', figsize=(10, 6))
plt.title('Top 10 Countries with Highest Immigration to Canada in 2013')
plt.xlabel('Country')
plt.ylabel('Number of Immigrants')
plt.xticks(rotation=90)
plt.show()
```

```python
# Pie Chart of Immigration by Development Status
immigration_dev_status = df.groupby('DevName')['Total'].sum()
immigration_dev_status.plot(kind='pie', figsize=(8, 8), autopct='%1.2f%%', startangle=90)
plt.title('Immigration to Canada by Development Status')
plt.ylabel('')
plt.show()
```

```python
# Histogram of Total Immigration
df['Total'].plot(kind='hist', bins=15, figsize=(10, 6))
plt.title('Histogram of Total Immigration to Canada')
plt.xlabel('Number of Immigrants')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Scatter Plot of Immigration in 2010 vs 2013
plt.figure(figsize=(10, 6))
plt.scatter(df['2010'], df['2013'], alpha=0.5)
plt.title('Immigration to Canada in 2010 vs 2013')
plt.xlabel('Immigration in 2010')
plt.ylabel('Immigration in 2013')
plt.grid(True)
plt.show()
```
